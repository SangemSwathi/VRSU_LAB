```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image in grayscale
img = cv2.imread("/content/cat_vrsu.jpg", 0)

# Check if image is loaded
if img is None:
    raise ValueError("Image not found")

# Apply binary thresholding
threshold_value, binary_img = cv2.threshold(
    img,
    thresh=100,
    maxval=255,
    type=cv2.THRESH_BINARY
)

# Display original and thresholded images
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.title("Original Grayscale Image")
plt.imshow(img, cmap='gray')
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Binary Thresholded Image")
plt.imshow(binary_img, cmap='gray')
plt.axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image in grayscale
img = cv2.imread("/content/cat_vrsu.jpg", 0)

# Check if image is loaded
if img is None:
    raise ValueError("Image not found")

# Apply Otsu's Thresholding
threshold_value, binary_img = cv2.threshold(
    img,
    23,                      # threshold value set to 0 for Otsu
    255,
    cv2.THRESH_BINARY + cv2.THRESH_OTSU
)

# Display original and thresholded images
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.title("Original Grayscale Image")
plt.imshow(img, cmap='gray')
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Otsu Thresholded Image")
plt.imshow(binary_img, cmap='gray')
plt.axis("off")

plt.show()

print("Otsu's computed threshold value:", threshold_value)
```
